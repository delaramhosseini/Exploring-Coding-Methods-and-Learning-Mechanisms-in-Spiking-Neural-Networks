```python
def STDPSmupation_PoissonDistribution(model=LIF(tau=10,
                        R=1.5,
                        u_rest=-67,
                        u_reset=-75,
                        threshold=-55,
                        ratio=1), 
                current=ConstanceCurrent(current=0), 
                dt = 1, 
                PLT = True,
                iteration_num = 1000,
                connection_pattern = FullConnectivitySecondOption(J0=10, standardـdeviation=30), 
                input_data = data1,
                positive_tau = 3,
                negative_tau = 2,
                A_negative = 5,
                A_positive = 5,
                normalize = True,
                input_time = 30,
                PostNG_size = 2
                ):
        
        PreNG_size = input_data.shape[1] 
                        
        net1 = pmt.Network(behavior = {1: Timeresolution(dt = dt)} )

        pre = pmt.NeuronGroup(size = PreNG_size,
                                net = net1,
                                behavior = {
                                        5 : current,
                                        6 : Dandrite(),
                                        7 : model,
                                        8 : PoissonDistribution(time = input_time, data=input_data, rest_time=5),
                                        9 : PopulationActivity(),
                                        13 : pmt.Recorder(variables=["u","I","w","T"],tag="ng1_rec"),
                                        14 : pmt.EventRecorder( variables=["spike"],tag="ng1_even"),
                                        })

        post = pmt.NeuronGroup(size = PostNG_size,
                                net = net1,
                                behavior = {
                                        5 : current,
                                        6 : Dandrite(),
                                        7 : LIF(tau=3,
                                                R=10,
                                                u_rest=-67,
                                                u_reset=-75,
                                                threshold=-55,
                                                ratio=1),
                                        9 : PopulationActivity(),
                                        13 : pmt.Recorder(variables=["u","I","w","T"],tag="ng1_rec"),
                                        14 : pmt.EventRecorder( variables=["spike"],tag="ng1_even"),
                                        })
        
        syn_group_from_pop1_to_pop2 = pmt.SynapseGroup(
                        tag="pre_to_post",
                        net=net1,
                        src=pre,
                        dst=post,
                        behavior={
                                4: connection_pattern,
                                10: STDP(positive_tau = positive_tau,
                                        negative_tau = negative_tau,
                                        A_negative = A_negative,
                                        A_positive = A_positive,
                                        normalize = normalize),
                                13: pmt.Recorder(variables=["W"], tag="pre_to_post"),
                                },
        )
        
        net1.initialize(info=False)
        net1.simulate_iterations(iteration_num)
        
        if PLT: plot( net1,
                # plot_title,
                ngs=[pre, post],
                sgs=[syn_group_from_pop1_to_pop2],
                print_sum_activities=True,
                scaling_factor= (3+ PreNG_size/40)
                )
        
        return net1
```
